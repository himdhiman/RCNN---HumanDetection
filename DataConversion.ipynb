```python
mapping = {}
img_list = os.listdir('./PennFudanPed/PennFudanPed/PNGImages/')
for img in img_list:
    img_name = img.split('.')[0]
    with open('./PennFudanPed/PennFudanPed/Annotation/'+img_name+'.txt', 'r') as f:
        text = f.readlines()
    l = []
    for i in range(5, len(text)):
        d = []
        if text[i].startswith("Bounding box for object"):
            data = text[i].split(":")[1].split("-")
            d.append(int(data[0].split(',')[0][2:]))
            d.append(int(data[0].split(',')[1][1:-2]))
            d.append(int(data[1].split(',')[0][2:]))
            d.append(int(data[1].split(',')[1][1:-2]))
            l.append(d)
    mapping[img_name] = l
```
